Na naszych zajęciach pisaliśmy program Hello World. Napiszmy tym razem program, który będzie witał użytkownika w wybranym przez niego (użytkownika) języku. Użytkownik będzie mógł wybrać język tylko spośród zaproponowanych przez program, a cały dialog ma być stosunkowo krótki, bo ma się zakończyć powitaniem wypisanym przez program. Początkowe propozycje program ma wypisywać po polsku. Program powinien "mówić" co najmniej czterema, dowolnie przez Państwa wybranymi językami. Przykład działania programu podano na końcu treści zadania.

Program powinien składać się z funkcji tekstPowitania, powitaj i wywołania funkcji powitaj.

Funkcja tekstPowitania ma mieć jeden parametr (kod wybranego języka, np. "kl"). Jej wynikiem ma być tekst powitania w tym języku (np. "nuqneH!"). Jeśli parametrem jest tekst nieodpowiadający kodowi żadnego ze znanych funkcji języków, wynikiem powinien być stosowny komunikat zawierający nierozpoznany kod języka.

Funkcja powitaj ma być bezparametrowa i nie ma dawać wyniku. W jej treści należy najpierw wypisać stosowne komunikaty dla użytkownika (p. przykład na końcu), potem wczytać wybór użytkownika, potem korzystając z funkcji tekstPowitania wygenerować powitanie i na koniec je wypisać. Czyli cały dialog z użytkownikiem ma się rozegrać wewnątrz funkcji powitaj.

Jako rozwiązanie należy przesłać plik o nazwie zgodnej ze schematem ImięNazwisko2.py (np. JanKowalski2.py). Plik należy wgrać do naszego kursu w Moodle'u (czyli tego kursu :) ).


Uwagi:

    W rozwiązaniu może się przydać instrukcja if z częścią elif (jej użycie nie jest wymagane, można zapisać rozwiązanie używając wyłącznie zagnieżdżonych instrukcji if).
    Języki nie muszą być europejskie, nie muszą być współczesne, ale muszą być prawdziwe (w sensie jak w przykładzie). Prosimy, aby Państwo podali w komentarzu dla każdego wybranego języka źródła w sieci, z jakich brali Państwo teksty powitań (np. można podać adres słownika internetowego dla wybranego języka, albo tekst rozmówek w tym języku).
    W programie można użyć języków podanych w przykładzie, ale liczę na Państwa pomysłowość!
    Opis źródeł tłumaczenia powitania w danym języku może wyglądać np. tak:
        # klingoński: Everyday Klingon Phrases (https://www.kli.org/about-klingon/klingon-phrases/)
    W razie wątpliwości można i należy zadawać pytania - ponieważ nasz kurs Moodle'owy dotyczy różnych grup, z różnymi zadaniami, akurat pytania o treści zadań domowych proszę zadawać n aforum dedykowanym naszej grupie lub mailowo.
    Treść funkcji zapewne będzie krótsza niż treść zadania - Python to fajny język :).
