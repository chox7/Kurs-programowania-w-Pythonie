## Zadanie małe (zajęcia 3 skrytptowy Python)

Zadanie jest grupowe - minimalny skład to Osoba 1 i Osoba 2.

Stwórz repozytorium w git-cie a w nim skrypt tworzący i czytającego pliki w następujący sposób. 

Skrypt, korzystając z biblioteki `argparse` ma pozwalać na określenie w parametrach następujących opcji:

a) wybór miesięcy (można wybrać dowolną ilość)

b) wybór *zakresu* dni tygodnia (tyle samo argumentów co w (a) )

c) wybór albo rano albo wieczorem (domyślnie rano) - dla ilości przypadków il.miesięcy x il.dni

Dla zadanych parametrów wygeneruj ścieżki (tu każdy dzień tygodnia z zakresu traktujemy osobno).

```Przykład:
Wejście: [styczeń, luty], [pn-wt, pt], [r,w]

Styczeń/poniedziałek/rano
Styczeń/wtorek/wieczorem
Luty/piątek/rano (bo domyślnie)
```

d) wybór jednoelementowej opcji np. -t (tworzenie) -u (usuwanie); lub jeden argument flaga do ustalenia w grupie 

e) wybór jednoelementowej opcji -c (csv) -j (json)

Następnie w każdej z wygenerownych ścieżek należy:

- utworzyć pliki zadanego typu

- wczytać pliki zadanego typu i podać wynik


**Osoba 1**

- tworzy repozytorium w git-cie i zarządza skryptem z najwyższego poziomu (obsługuje parametry)

**Osoba 2**

- tworzy część generującą ścieżki i tworzącą pliki (na tych ścieżkach) lub odczytującą informację z takich plików

**Osoba 3 / Osoba 1**

- tworzy zapis do pliku csv

The 'Solutions.csv' should consist of two lines:
" Model; Output value; Time of computation; "
" A ; 17 ; 465s; "

The first line is always the same.
The second one should be generated randomly (x in {A,B,C}, 0-1000, 0-1000s)

- tworzy odczyt z pliku csv

Output the sum of "Time of computation" for the model A.
Assume that you do not know how many folders and subfolders there are.

**Osoba 4 / Osoba 2**
- odczyt i zapis do pliku json z analogicznymi wymaganiami jak dla csv
